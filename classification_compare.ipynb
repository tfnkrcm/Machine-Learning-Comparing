```python
# Gerekli Kütüphanelerin Yüklenmesi
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Örnek olarak basit bir sınıflandırma problemi için sentetik bir veri seti oluşturalım.
X, y = make_classification(n_samples=1000, n_features=20, n_informative=2, n_redundant=10, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Modellerin Eğitilmesi ve Tahmin Yapılması
# MLP Classifier
mlp = MLPClassifier(random_state=42)
mlp.fit(X_train, y_train)
mlp_preds = mlp.predict(X_test)

# SVM Classifier
svm = SVC(random_state=42)
svm.fit(X_train, y_train)
svm_preds = svm.predict(X_test)

# Random Forest Classifier
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)

# Performansın Değerlendirilmesi
# Accuracy puanlari
mlp_accuracy = accuracy_score(y_test, mlp_preds)
svm_accuracy = accuracy_score(y_test, svm_preds)
rf_accuracy = accuracy_score(y_test, rf_preds)

print(f"MLP Accuracy: {mlp_accuracy}")
print(f"SVM Accuracy: {svm_accuracy}")
print(f"Random Forest Accuracy: {rf_accuracy}")

# Sonuçların Grafik Üzerinde Karşılaştırılması
algorithms = ['MLP', 'SVM', 'Random Forest']
accuracies = [mlp_accuracy, svm_accuracy, rf_accuracy]

plt.figure(figsize=(10, 6))
plt.bar(algorithms, accuracies, color=['blue', 'green', 'red'])
plt.xlabel('Algorithms')
plt.ylabel('Accuracy')
plt.title('Comparison of Machine Learning Algorithms')
plt.ylim([0, 1])
plt.show()
```
